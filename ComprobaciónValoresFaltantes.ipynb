```python
# Importamos las librerías necesarias
import pandas as pd
import os

# Ruta donde están las bases de datos después de aplicar MissForest
ruta = r"C:\Users\guill\TFG\BASESDEDATOS\MissForest\todos"

# Obtenemos todos los archivos Excel que hay en la carpeta
archivos = [f for f in os.listdir(ruta) if f.endswith(".xlsx")]

# Creamos una lista vacía donde se guardarán los resultados
resultados = []

# Recorremos cada archivo Excel de la carpeta
for archivo in archivos:
    
    # Construimos la ruta completa del archivo
    path = os.path.join(ruta, archivo)
    
    # Leemos el archivo Excel
    df = pd.read_excel(path)
    
    # Calculamos el número total de celdas de la base
    total_celdas = df.shape[0] * df.shape[1]
    
    # Calculamos el número total de valores faltantes
    missing = df.isna().sum().sum()
    
    # Calculamos el porcentaje de valores faltantes
    porcentaje = (missing / total_celdas) * 100
    
    # Guardamos los resultados de la base actual
    resultados.append({
        "Base": archivo,
        "Total celdas": total_celdas,
        "Valores faltantes": missing,
        "% valores faltantes": round(porcentaje, 2)
    })

# Convertimos la lista de resultados en una tabla
tabla_missing_missforest = pd.DataFrame(resultados)

# Mostramos la tabla final
print(tabla_missing_missforest)
```
